```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.title("Simple Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y, color="cyan",linestyle="dashed",marker="*",linewidth=2)
plt.title("Simple Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x=["Jan","Feb","March","April","May","June"]
y=[1000,2000,4000,6000,8000,1000]
plt.plot(x,y, color="green",linestyle="dashed",marker="o",linewidth=3)
plt.title("Growth of savings amount")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color='Skyblue',linewidth=10)
plt.title("Basic Bar Chart")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color=['cyan','lavender','skyblue','yellow','violet'],edgecolor="black",width=1)
plt.title("Basic Bar Chart")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
ages=[20,21,22,33,34,52,23,43,23,43,12,31]
b=[10,20,30,40,50,60]
#bins defines the intervals for grouping the data
plt.hist(ages,bins=b,color='pink',edgecolor="purple")
plt.title("Age distribution of customers")
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
sizes=[40.0,10.0,20.0,30.0]
labels=['Python','Ruby','c++','java']
colors=['cyan','lavender','lightgreen','violet']
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',startangle=90)
plt.title("Basic Pie Chart")
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[23,21,12,3,4,5]
plt.scatter(x,y,color="green",s=100,edgecolor="black")
plt.title("Simple scatter plot")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
sales_analysis=pd.read_csv("sales_analysis.csv")
sales_analysis.head()
```

```python
#Use line plots to track monthly sales 
plt.figure(figsize=(12, 6))  # Increase figure size
plt.plot(
    sales_analysis["Month"], 
    sales_analysis["Sales"], 
    color="skyblue", 
    linestyle="dashed", 
    marker="o", 
    linewidth=3
)
plt.title("Monthly Sales Tracking", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Sales", fontsize=12)
plt.xticks(rotation=45)  
plt.grid(alpha=0.5)  
plt.show()
```

```python
#histograms for order quantity distribution.
plt.hist(
    sales_analysis['OrderQuantity'], 
    bins=10, 
    color='purple', 
    edgecolor="black"
)
plt.title("Order Quantity Distribution", fontsize=16)
plt.xlabel("Order Quantity Ranges", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
customer=pd.read_csv("customer_demographics.csv")
customer.head()
```

```python
#Customer Demographics: Visualize age distribution (histogram) 
plt.figure(figsize=(8, 6)) 
plt.hist(customer['Age'], bins=10, color='lavender', edgecolor="purple")
plt.title("Age Distribution of Customers", fontsize=16)
plt.xlabel("Age Ranges", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#spending patterns(scatter plot).
plt.scatter(customer['Age'], customer['Spending'], color='blue', alpha=0.7)
plt.title("Spending Patterns by Age", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Spending ($)", fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
school_df=pd.read_csv("school_performance.csv")
school_df.head()
```

```python
#Compare students' marks across subjects (bar chart) 
subjects = ['Math', 'Science', 'English']
marks = school_df[subjects].mean() 
plt.bar(subjects, marks, color='Skyblue', linewidth=10)
plt.title("Average Marks Across Subjects")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.show()
```

```python
#identify outliers (box plot).
plt.boxplot([school_df['Math'], school_df['Science'], school_df['English']], labels=subjects)
plt.title("Marks Distribution and Outliers (Box Plot)")
plt.ylabel("Marks")
plt.show()
```
